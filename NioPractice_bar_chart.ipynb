```python
import pandas as pd

df = pd.read_csv('/content/NioPracticeUserbase.csv') #Read Data & convert it to the dataframe

df.head() #show 1st 5 rows of the data frame
```

```python
import plotly.express as px
import pandas as pd

fig = px.bar(df, x='Gender', color='Gender', title='User Demographics by Gender')
fig.show()
```

```python
from dash import Dash, dcc, html
from dash.dependencies import Input, Output
import plotly.express as px
import pandas as pd

app = Dash(__name__)

app.layout = html.Div([
    dcc.Dropdown(
        id='country-dropdown',
        options=[{'label': i, 'value': i} for i in df['Country'].unique()],
        value=df['Country'].unique()[0], # Default value
        clearable=False,
    ),
    dcc.Graph(id='gender-bar-chart'),
])

@app.callback(
    Output('gender-bar-chart', 'figure'),
    [Input('country-dropdown', 'value')]
)
def update_chart(selected_country):
    filtered_df = df[df['Country'] == selected_country]
    fig = px.bar(filtered_df, x='Gender', color='Gender',
                 title=f'User Demographics by Gender for {selected_country}')
    return fig

if __name__ == '__main__':
    app.run_server(debug=True)
```
